```python
##IMPORTAMOS LAS LIBRERIAS PANDAS Y MATPLOTLIB PARA LAS GRAFICAS
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
datos = pd.read_csv('abnerlee_temperatura.csv',header = 0)
df = pd.DataFrame(datos)
#MOSTRAMOS LOS DATOS DE NUESTRO CSV
print (datos)
```

```python
##GRAFICAS EN LINEA O DE BARRAS
```

```python
df.groupby('Temperatura')['Lluvia'].sum().plot(kind='bar')
```

```python
df.groupby('Lluvia')['Lluvia'].sum().plot(kind='bar')
```

```python
df.groupby('Humedad')['Lluvia'].sum().plot(kind='bar')
```

```python
#GRAFICAS DE PASTEL
```

```python
df.groupby('Temperatura')['Lluvia'].sum().plot(kind='pie')
```

```python
df.groupby('Lluvia')['Lluvia'].sum().plot(kind='pie')
```

```python
df.groupby('Humedad')['Lluvia'].sum().plot(kind='pie')
```

```python
df.Temperatura.groupby(df.Humedad).sum().plot(kind='pie',cmap='Paired')
plt.axis('equal')
```
